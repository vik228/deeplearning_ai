#### Broadcasting in Numpy

> __During artihmatic operations, the smaller array is broadcasted towards the larger array, so as to make the shapes compatible. Following are the rules of broadcasting__
> 1. __When the two arrays are equal__
> 2. __When one is array NxN and other is 1X1__
> 3. __When one is array NxN and other is Nx1__
> 4. __When one is array NxN and other is 1XN__

1. When one is NxN and other is 1X1

```python
z = np.arange(9).reshape(3,3)
```

```python
z
```

```python
z = z + 1
```

```python
z.shape
```

```python
z
```

2. When one is NxN and other is Nx1

```python
Z1 = np.arange(9).reshape(3,3)
Z2 = np.arange(3)[::-1].reshape(3,1)
```

```python
Z3 = Z1 + Z2
```

3. When one is NxN and other is 1XN

```python
Z1 = np.arange(9).reshape(3,3)
Z2 = np.arange(3)[::-1]
```

```python
Z3 = Z1 + Z2
```
