```python
# Query to find the stories posted on jan 1 2012 along with the number of comments
query = """
    WITH stories AS 
( 
           SELECT     s.text                                      AS stories_text, 
                      c.id                                        AS c_id 
           FROM       `bigquery-public-data.hacker_news.stories`  AS s 
           INNER JOIN `bigquery-public-data.hacker_news.comments` AS c 
           ON         s.id = c.parent 
           where Extract(date FROM s.time_ts) = '2012-01-01') 
SELECT   stories.stories_text, 
         count(1) AS num_comments 
FROM     stories  
GROUP BY stories.stories_text 
ORDER BY num_comments desc
"""

bigquery.run_query(query, maximum_bytes_billed=10**10).to_dataframe()
```

```python
# Query to select users who posted comments or stories on 1 jan 2014
query = """
    select c.by from `bigquery-public-data.hacker_news.comments` AS c 
    where extract(DATE from c.time_ts) = '2014-01-01'
    UNION DISTINCT 
    select s.by from `bigquery-public-data.hacker_news.stories` AS s 
    where extract(DATE from s.time_ts) = '2014-01-01'
"""
bigquery.run_query(query, maximum_bytes_billed=10**10).to_dataframe()
```

```python
bikeshare_trips_table = bigquery_sf.get_table("bikeshare_trips")
bigquery.list_rows(bikeshare_trips_table, max_results=5)
```

```python
# query to find the cumulative trips.
query = """
    with data as (
        select 
            extract(date from start_date) as trip_date,
            count(*) as num_trips
        from 
            `bigquery-public-data.san_francisco.bikeshare_trips`
        where 
            extract(year from start_date) = 2015
        group by
            trip_date
        
    ) select *,
             sum(num_trips)
             over(
                 order by trip_date
                 ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
             ) as cumulative_trips
      from 
          data
"""
bigquery.run_query(query, maximum_bytes_billed=10**10).to_dataframe()
```

```python
# query tracks the stations where each bike began (in start_station_id) and ended (in end_station_id) the day on October 25, 2015.

query = """
    select 
           time(start_date) as trip_time,
           bike_number,
           first_value(start_station_id)
               over(
                   partition by bike_number
                   order by start_date
                   ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
               ) as first_station_id,
           last_value(end_station_id)
               over(
                   partition by bike_number
                   order by start_date
                   ROWS BETWEEN UNBOUNDED PRECEDING AND UNBOUNDED FOLLOWING
               ) as last_station_id,
           start_station_id,
           end_station_id
    FROM 
        `bigquery-public-data.san_francisco.bikeshare_trips`
    WHERE 
        DATE(start_date) = '2015-10-25'
           
"""
bigquery.run_query(query, maximum_bytes_billed=10**10).to_dataframe()
```
